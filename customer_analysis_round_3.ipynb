```python
import pandas as pd
df = pd.read_csv('./files_for_lab/csv_files/marketing_customer_analysis.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# - Show a plot of the response rate by the sales channel.
response_by_channel = df.groupby('Sales Channel')['Response'].value_counts(normalize=True).unstack().fillna(0)
response_rate_yes = response_by_channel.get('Yes', pd.Series(0, index=response_by_channel.index)).plot(kind='bar', title='Response Rate by Sales Channel', ylabel='Response Rate (Yes)', xlabel='Sales Channel')
```

```python
response_by_claim = df.groupby('Total Claim Amount')['Response'].value_counts(normalize=True).unstack().fillna(0)
response_rate_yes_claim = response_by_claim.get('Yes', pd.Series(0, index=response_by_claim.index)).plot(kind='bar', title='Response Rate by Total Claim Amount', ylabel='Response Rate (Yes)', xlabel='Total Claim Amount')    
```

```python
response_by_income = df.groupby('Income')['Response'].value_counts(normalize=True).unstack().fillna(0)
response_rate_yes_income = response_by_income.get('Yes', pd.Series(0, index=response_by_income.index)).plot(kind='bar', title='Response Rate by Income', ylabel='Response Rate (Yes)', xlabel='Income')
```
